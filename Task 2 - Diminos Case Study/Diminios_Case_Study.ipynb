Import required libraries

```python
import pandas as pd
import numpy as np
```

Load the dataset

```python
df = pd.read_csv("diminos_data.csv")
```

Convert timestamp columns to datetime

```python
df["order_placed_at"] = pd.to_datetime(df["order_placed_at"])
df["order_delivered_at"] = pd.to_datetime(df["order_delivered_at"])
```

Calculate delivery time in minutes

```python
df["delivery_time"] = (df["order_delivered_at"] - df["order_placed_at"]).dt.total_seconds() / 60
```

Compute 95th percentile delivery time

```python
p95_delivery_time = np.percentile(df["delivery_time"], 95)
print("95th Percentile Delivery Time:", round(p95_delivery_time, 2), "minutes")
```

Check Dimino’s SLA compliance

```python
if p95_delivery_time < 31:
    print("Store meets Dimino’s delivery SLA.")
else:
    print("Store does NOT meet Dimino’s delivery SLA.")
```

Summary statistics

```python
print("Average Delivery Time:", round(df["delivery_time"].mean(), 2))
print("Minimum Delivery Time:", df["delivery_time"].min())
print("Maximum Delivery Time:", df["delivery_time"].max())
```
