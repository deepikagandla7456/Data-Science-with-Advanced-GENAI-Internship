```python
start_times = [[1, 2, 3],[4]]
end_times = [[3, 2, 7],[4]]
query_times = [4, 4]
for i in range(len(start_times)):
    count = 0
    startTime = start_times[i]
    endTime = end_times[i]
    queryTime = query_times[i]
    for j in range(len(startTime)):
        if startTime[j] <= queryTime <= endTime[j]:
            count += 1
    print(count)
```
