```python
ratings = [[2, 5, 3, 4, 1],[2, 1, 3],[1, 2, 3, 4]]
for rating in ratings:
    count = 0
    n = len(rating)
    for i in range(n):
        for j in range(i + 1, n):
            for k in range(j + 1, n):
                if rating[i] < rating[j] < rating[k]:
                    count += 1
                elif rating[i] > rating[j] > rating[k]:
                    count += 1
    print(count)
```
