```python
class Solution:
    def subtractProductAndSum(self, n):
        product = 1
        total_sum = 0  
        for digit in str(n):
            value = int(digit)
            product *= value
            total_sum += value 
        return product - total_sum
if __name__ == "__main__":
    s = Solution() 
    input1 = 234
    print(s.subtractProductAndSum(input1))   
    input2 = 4421
    print(s.subtractProductAndSum(input2))
```
