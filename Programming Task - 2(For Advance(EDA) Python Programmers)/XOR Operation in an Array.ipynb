```python
class Solution:
    def xorOperation(self, n, start):
        xor_value = 0   
        for i in range(n):
            current = start + 2 * i
            xor_value = xor_value ^ current           
        return xor_value
if __name__ == "__main__":
    s = Solution()
    input1_n = 5
    input1_start = 0
    print(s.xorOperation(input1_n, input1_start)) 
    input2_n = 4
    input2_start = 3
    print(s.xorOperation(input2_n, input2_start))
```
