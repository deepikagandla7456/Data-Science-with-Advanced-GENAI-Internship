```python
class Solution:
    def findNumbers(self, nums):
        count = 0
        for num in nums:
            digits = len(str(num))
            if digits % 2 == 0:
                count += 1
        return count
if __name__ == "__main__":
    s = Solution()
    input = [12,345,2,6,7896]
    output = s.findNumbers(input)
    print(output)
    
    input1 = [555,901,482,1771]
    output1 = s.findNumbers(input1)
    print(output1)
```
