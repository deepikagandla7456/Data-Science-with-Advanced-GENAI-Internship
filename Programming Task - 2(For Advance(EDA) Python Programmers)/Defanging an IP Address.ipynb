```python
class Solution:
    def defangIPaddr(self, address):
        defanged = address.replace('.', '[.]')
        return defanged
if __name__ == "__main__":
    s = Solution()
    
    input = "1.1.1.1"
    output = s.defangIPaddr(input)
    print('"'+output+'"')
    input1 = "255.100.50.0"
    output1 = s.defangIPaddr(input1)
    print('"' + output1 + '"')
```
