```python
class Solution:
    def smallerNumbersThanCurrent(self, nums):
        result = []
        
        for i in range(len(nums)):
            count = 0
            for j in range(len(nums)):
                if nums[j] < nums[i]:
                    count += 1
            result.append(count)
        return result
if __name__ == "__main__":
    s = Solution() 
    input1 = [8, 1, 2, 2, 3]
    print(s.smallerNumbersThanCurrent(input1)) 
    input2 = [6, 5, 4, 8]
    print(s.smallerNumbersThanCurrent(input2))
    input3 = [7,7,7,7]
    print(s.smallerNumbersThanCurrent(input3))
```
