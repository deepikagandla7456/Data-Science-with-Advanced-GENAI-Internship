```python
class Solution:
    def numIdenticalPairs(self, nums):
        count = 0
        n = len(nums)       
        for i in range(n):
            for j in range(i + 1, n):
                if nums[i] == nums[j]:
                    count += 1            
        return count
if __name__ == "__main__":
    s = Solution()    
    input1 = [1, 2, 3, 1, 1, 3]
    print(s.numIdenticalPairs(input1))
    input2 = [1, 1, 1, 1]
    print(s.numIdenticalPairs(input2))
    input3 = [1,2,3]
    print(s.numIdenticalPairs(input3))
```
